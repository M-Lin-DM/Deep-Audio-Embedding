```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = "wav-file-name"

architecture_ID = "R-F"

Delta_path_RF = 'C:/Users/MrLin/Documents/Experiments/Deep Audio Embedding/' + dataset + '/variables/Delta_' + architecture_ID

architecture_ID = "R"
Delta_path_R = 'C:/Users/MrLin/Documents/Experiments/Deep Audio Embedding/' + dataset + '/variables/Delta_' + architecture_ID
```

```python
varbs = np.load(Delta_path_RF + '.npz')
varbs.files
Delta_mean_RF = varbs['Delta_mean']
Delta_std_RF = varbs['Delta_std']

varbs = np.load(Delta_path_R + '.npz')
Delta_mean_R = varbs['Delta_mean']
Delta_std_R = varbs['Delta_std']

N_steps = 30
```

```python
fig_save_path = 'C:/Users/MrLin/Documents/Experiments/Deep Audio Embedding/' + dataset + '/Results/'

a=7
fig2 = plt.figure(figsize=(1.7778*a, a))  #  e.g. figsize=(4, 3) --> img saved has resolution (400, 300) width by height when using dpi='figure' in savefig

ax = fig2.gca()
plt.errorbar(np.arange(0, N_steps, 1), Delta_mean_R, yerr=Delta_std_R/np.sqrt(806), xerr=None, fmt='k', capsize=3, label ='Reconstruction loss')
plt.errorbar(np.arange(0, N_steps, 1), Delta_mean_RF, yerr=Delta_std_RF/np.sqrt(806), xerr=None, fmt='r', capsize=3, label ='Reconstruction+Forecasting loss')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
# plt.ylabel('distance from true path')
# plt.xlabel('number of forecast steps')
ax.set_xlabel('number of forecast steps', fontsize = 15)
ax.set_ylabel("distance from forecasted path to true path", fontsize = 15)
plt.legend(fontsize=15)
plt.show()
fig2.savefig(fig_save_path + 'Delta.png', transparent=False, dpi='figure', bbox_inches=None)
```
